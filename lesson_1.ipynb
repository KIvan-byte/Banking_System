```python
def dec_to_bin(n):
  if n == 0:
    return "0"

  bin_num = ""
  while n > 0:
    bin_num += str(n % 2)
    n //= 2

  return bin_num[::-1]


def main():
  n = int(input("Введите десятичное число: "))
  bin_num = dec_to_bin(n)
  print(f"Двоичное число: {bin_num}")


if __name__ == "__main__":
  main()
```

```python
def reverse_yes_no(sentence):
  reversed_sentence = []
  for word in sentence.split():
    if word == "yes":
      reversed_sentence.append("no")
    elif word == "no":
      reversed_sentence.append("yes")
    else:
      reversed_sentence.append(word)

  return " ".join(reversed_sentence)


def main():
  sentence = input("Enter a sentence: ")
  print(f"The modified sentence is: {reverse_yes_no(sentence)}")


if __name__ == "__main__":
  main()
```

```python
def add_element_to_dictionary(dictionary, key, value):
  dictionary[key] = value


def merge_dictionaries(dictionary1, dictionary2, dictionary3):
  merged_dictionary = dictionary1.copy()
  merged_dictionary.update(dictionary2)
  merged_dictionary.update(dictionary3)
  return merged_dictionary


def check_key_exists(dictionary, key):
  return key in dictionary


def generate_dictionary(n):
  dictionary = {}
  for i in range(1, n + 1):
    dictionary[i] = i ** 2
  return dictionary


def find_maximum_integer_value(dictionary):
  max_value = None
  for value in dictionary.values():
    if max_value is None or value > max_value:
      max_value = value
  return max_value


def sum_dictionary_values(dictionary):
  sum_of_values = 0
  for value in dictionary.values():
    sum_of_values += value
  return sum_of_values


def main():
  dictionary = {"key1": "value1", "key2": "value2"}
  add_element_to_dictionary(dictionary, "key3", "value3")
  print(dictionary)

  dictionary1 = {"key1": "value1", "key2": "value2"}
  dictionary2 = {"key3": "value3", "key4": "value4"}
  dictionary3 = {"key5": "value5", "key6": "value6"}
  merged_dictionary = merge_dictionaries(dictionary1, dictionary2, dictionary3)
  print(merged_dictionary)

  print(check_key_exists(dictionary, "key1"))
  print(check_key_exists(dictionary, "key3"))

  n = int(input("Введите максимальное значение ключа: "))
  dictionary = generate_dictionary(n)
  print(dictionary)

  print(find_maximum_integer_value(dictionary))

  print(sum_dictionary_values(dictionary))


if __name__ == "__main__":
  main()
```
